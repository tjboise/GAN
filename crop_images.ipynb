#### Crop the road top down view images ($1014 * 2011$) into $256 * 256$ sizes.

```python
from PIL import Image
import random

im=Image.open("C:/Users/tjzha/OneDrive/Documents/GitHub/GAN/data/road_image/0 (1).JPG")
```

```python
im
```

```python
m,n=im.size
w=256
h=256

for i in range(20):
    x = random.randint(0,m-w)
    y = random.randint(0,n-h)
    region = im.crop((x,y,x+w,y+h))
    region.save("C:/Users/tjzha/OneDrive/Documents/GitHub/GAN/data/road_image/crop_image/"+str(i)+".jpg")
```

process a lot of images.

```python
import os

file_root = "C:/Users/tjzha/OneDrive/Documents/GitHub/GAN/data/road_image"
file_list = os.listdir(file_root)
print(file_list)

w=256
h=256
save_out = "C:/Users/tjzha/OneDrive/Documents/GitHub/GAN/data/road_crop_image/"
for img_name in file_list:
    img_path = file_root+'/'+img_name
    im=Image.open(img_path)
    m,n=im.size
    for i in range(20):
        x = random.randint(0,m-w)
        y = random.randint(0,n-h)
        region = im.crop((x,y,x+w,y+h))
        region.save(save_out+str(img_name[:-4])+'-'+str(i)+".jpg")
    
    
    
```
