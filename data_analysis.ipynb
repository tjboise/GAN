```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

unet1 = pd.read_csv('Unet.csv')
unet2 = pd.read_csv('Unet2.csv')
FCN1 = pd.read_csv('FCN.csv')
FCN2 = pd.read_csv('FCN2.csv')
deeplab1 = pd.read_csv('deeplab.csv')
deeplab2 = pd.read_csv('deeplab2.csv')
LRASPP1 = pd.read_csv('LRASPP.csv')
LRASPP2 = pd.read_csv('LRASPP2.csv')
Attunet1 = pd.read_csv('Attunet.csv')
Attunet2 = pd.read_csv('attunet2.csv')
Attunetmin1 = pd.read_csv('attunetmin1.csv')
Attunetmin2 = pd.read_csv('attunetmin2.csv')

unet1['loss2']=unet2['loss']
unet1['mean']=(unet1['loss']+unet1['loss2'])/2
unet1['std']=np.std(unet1[['loss','loss2']],axis=1)

FCN1['loss2']=FCN2['loss']
FCN1['mean']=(FCN1['loss']+FCN1['loss2'])/2
FCN1['std']=np.std(FCN1[['loss','loss2']],axis=1)

deeplab1['loss2']=deeplab2['loss']
deeplab1['mean']=(deeplab1['loss']+deeplab1['loss2'])/2
deeplab1['std']=np.std(deeplab1[['loss','loss2']],axis=1)


LRASPP1['loss2']=LRASPP2['loss']
LRASPP1['mean']=(LRASPP1['loss']+LRASPP1['loss2'])/2
LRASPP1['std']=np.std(LRASPP1[['loss','loss2']],axis=1)


Attunet1['loss2']=Attunet2['loss']
Attunet1['mean']=(Attunet1['loss']+Attunet1['loss2'])/2
Attunet1['std']=np.std(Attunet1[['loss','loss2']],axis=1)


Attunetmin1['loss2']=Attunetmin2['loss']
Attunetmin1['mean']=(Attunetmin1['loss']+Attunetmin1['loss2'])/2
Attunetmin1['std']=np.std(Attunetmin1[['loss','loss2']],axis=1)


LRASPP1
```

```python
plt.style.use('classic')
plt.plot(unet1['mean'],label='U-Net',color='blue')

y1=unet1['mean']-unet1['std']
y2=unet1['mean']+unet1['std']
plt.fill_between(unet1['Unnamed: 0'],y1,y2
                 ,color='blue',alpha=0.3)



plt.plot(FCN1['loss'],label='FCN',color='red')

y1=FCN1['mean']-FCN1['std']
y2=FCN1['mean']+FCN1['std']
plt.fill_between(FCN1['Unnamed: 0'],y1,y2
                 ,color='red',alpha=0.3)


plt.plot(deeplab1['loss'],label='DeepLabv3',color='green')

y1=deeplab1['mean']-deeplab1['std']
y2=deeplab1['mean']+deeplab1['std']
plt.fill_between(deeplab1['Unnamed: 0'],y1,y2
                 ,color='green',alpha=0.3)


plt.plot(LRASPP1['loss'],label='LRASPP',color='orange')

y1=LRASPP1['mean']-LRASPP1['std']
y2=LRASPP1['mean']+LRASPP1['std']
plt.fill_between(LRASPP1['Unnamed: 0'],y1,y2
                 ,color='orange',alpha=0.3)


plt.plot(Attunet1['loss'],label='AttuNet',color='c')

y1=Attunet1['mean']-Attunet1['std']
y2=Attunet1['mean']+Attunet1['std']
plt.fill_between(Attunet1['Unnamed: 0'],y1,y2
                 ,color='c',alpha=0.3)

plt.plot(Attunetmin1['loss'],label='AttuNet_min',color='m')

y1=Attunetmin1['mean']-Attunetmin1['std']
y2=Attunetmin1['mean']+Attunetmin1['std']
plt.fill_between(Attunetmin1['Unnamed: 0'],y1,y2
                 ,color='m',alpha=0.3)

# plt.plot(Attunet['loss'],label='Attunet')
# plt.plot(Attunet['loss'],label='Attunet')



plt.yscale('log')
plt.ylabel('Loss')
plt.xlabel('Epochs')
#plt.xscale('log')
plt.legend()

plt.savefig('training.png',dpi=1200,bbox_inches = 'tight')
```
